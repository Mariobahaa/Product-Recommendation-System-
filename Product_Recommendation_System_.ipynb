```
import pandas as pd
import gzip
import json
import numpy as np

# convert from json to pandas
def parse(path):
  g = gzip.open(path, 'rb')
  for l in g:
    yield json.loads(l)

def getDF(path):
  i = 0
  df = {}
  for d in parse(path):
    df[i] = d
    i += 1
  return pd.DataFrame.from_dict(df, orient='index')

# make dict and fit it by ids
def dict_ids(col):
  id = 1 # 1481  0
  dic = dict() 
  for i in col:
    if i in dic:
      dic[i] = dic.get(i)
    else:
      dic[i] = id
      id += 1
  return dic, (id-1)

# make new col by mapping last ids to new ids
def new_col(col,dic):
  lis = list()
  for i in col:
    #if i in dic:
    lis.append(dic.get(i))
  return lis

#create products per user map
def productsOfUser(userId, productId, ratings):
  productsOfUserMap = dict()
  ratesOfUserMap = dict()
  len_of_products = []

  for i in range(0,len(userId)):
    productsOfUserMap[userId[i]] = np.array([]).astype(int) #intm
    ratesOfUserMap[userId[i]] = np.array([]).astype(int) #intm

  for i in range(0,len(userId)): 
    if productId[i] > -1:
      productsOfUserMap[userId[i]] = np.append(productsOfUserMap[userId[i]] ,productId[i])
      ratesOfUserMap[userId[i]] = np.append(ratesOfUserMap[userId[i]] ,ratings[i])
  
  for i in productsOfUserMap.values():
    len_of_products.append((len(i)-1))
  len_of_products = np.array(len_of_products)
  return productsOfUserMap,len_of_products,ratesOfUserMap

def fixed_product_list(prodOfUser,len_of_products,numOfProds):
  # for i in range(0,len(len_of_products)):
  #   if max(len_of_products) != len_of_products[i]:
  #     zeroes= np.zeros(max(len_of_products) - len_of_products[i])
  #     prodOfUser[i] = np.append(prodOfUser[i],zeroes)
  zeroes = np.ones((len(len_of_products),max(len_of_products))) * numOfProds 
  for  i in range(0,len(prodOfUser)):
    temp = np.array(prodOfUser[i])
    zeroes[i,0:len(temp)] = temp
  return zeroes , max(len_of_products)

def duplicateUserBehaviour(X):
    newlist = []
    tests = []
    for item in X:
        userId = item[0]
        behaviour = item[1]
        rating = item[2]
        lastIndex = -1

        for i in range (0,len(behaviour)-1):
            itemList = []
            behaviourList = []
            itemList.append(userId)
            for j in range(0,i+1):
                behaviourList.append(behaviour[j])
                lastIndex = j
            candidate = behaviour[lastIndex+1]
            canrating = rating[lastIndex+1]
            if i == len(behaviour)-2:
                itemList.append(behaviourList)
                itemList.append(candidate)
                itemList.append(canrating)
                tests.append(itemList)
            else:
                itemList.append(behaviourList)
                itemList.append(candidate)
                itemList.append(canrating)
                newlist.append(itemList)
    return newlist, tests
```

```
from tensorflow.python.keras.models import Model
import numpy as np
from tensorflow.python.keras.models import Sequential, load_model
from tensorflow.python.keras.layers import  AveragePooling1D, Flatten, Dense, Dropout, Embedding, Lambda, Input, Concatenate, PReLU, BatchNormalization
from tensorflow.python.keras.callbacks import ModelCheckpoint
from tensorflow.python.keras.metrics import binary_accuracy , categorical_accuracy
from tensorflow.python.keras import regularizers
from tensorflow.python.keras.initializers import TruncatedNormal
import tensorflow as tf

embedOutDim = 400 #200,400,1000
attention_scale = 100 #1,10,100
poolSize = 2

#, embeddings_regularizer= regularizers.l2(0.01) 1485

##################################################################
userInput =  Input(shape=(1,), name= 'users')
#userInput =  tf.keras.layers.Masking(mask_value=0.0)(userInput)
userEmbed =  Embedding(input_dim =numberOfUsers+1,output_dim= embedOutDim, mask_zero =True) (userInput)


##################################################################
prodInput =  Input(shape=(maxLen,), name= 'products') #batch_shape(None, None)
#prodInputs =  tf.keras.layers.Masking(mask_value=0)(prodInput)
prodEmbed =  Embedding(input_dim =numProds,output_dim= embedOutDim, mask_zero =True ) (prodInput)
#print(prodInput.shape)

##################################################################
# cateInput = Input(shape=(11,), name= 'categories')
# cateEmbed =  Embedding(input_dim = numberOfCategories , output_dim=embedOutDim, embeddings_regularizer= regularizers.l2(0.01)) (cateInput)

##################################################################
candProdInput = Input(shape=(1,),  name= 'candidateProdId')
candProdEmbed =  Embedding(input_dim = numProds +1 , output_dim=embedOutDim, mask_zero =True) (candProdInput)

##################################################################
# candCateInput = Input(shape=(1,),  name= 'candidateCateId')
# candCateEmbed =  Embedding(input_dim = numberOfCategories , output_dim=embedOutDim, embeddings_regularizer= regularizers.l2(0.01)) (candCateInput)

#lenInput = Input(shape=(1,), name= 'len_of_products')

candProdEmbed = Dropout(0.5)(candProdEmbed)
prodEmbed = Dropout(0.5)(prodEmbed)
userEmbed = Dropout(0.5)(userEmbed)


#Goods  =Concatenate()([prodEmbed,cateEmbed]) #Uncomment for original work 
Goods = prodEmbed

#Goods  =  Lambda(lambda x: tf.keras.backend.batch_flatten(x))(Goods)
#print(Goods.shape)
# SeqPoolLayer = SequencePoolingLayer("sum",supports_masking= False)([Goods,lenInput]) #Error 

WeightedGoods = tf.keras.layers.Attention(use_scale= attention_scale)([Goods,candProdEmbed])

#WeightedGoods = tf.convert_to_tensor(WeightedGoods)
#print(Goods.shape)
#print(WeightedGoods.shape)
SeqPoolLayer = tf.reduce_sum(WeightedGoods,axis=1,keep_dims=True, name='pooling') #First Parameter was "Goods"
print(SeqPoolLayer.shape)
#SeqPoolLayer = tf.keras.backend.sum(Goods, axis=1, keepdims=True)

#Candidate = Concatenate()([candProdEmbed,candCateEmbed]) #Uncomment for original work
Candidate = candProdEmbed

#Candidate = tf.reshape(Candidate, (tf.shape(Candidate)[0],1, tf.shape(Candidate)[1]*tf.shape(Candidate)[2])) 
#Candidate = tf.reshape(Candidate,(Candidate.shape[0],1,Candidate.shape[1]*Candidate.shape[2]))

Concatendated = Concatenate()([ userEmbed, SeqPoolLayer, Candidate])
#print(Concatendated.shape)

#Flattened     = Lambda(lambda x: tf.keras.backend.batch_flatten(x))(Concatendated)
Flattened = Flatten()(Concatendated)
#print(Flattened.shape)

#BN = BatchNormalization()(Flattened)
#tf.convert_to_tensor(Flattened)

Dense1 =Dense(200) (Flattened)
Dense1 =Dropout(0.5) (Dense1)
PDense1 = PReLU() (Dense1)

#, kernel_initializer=TruncatedNormal() 1486

Dense2 = Dense(80)(PDense1)
Dense2 =Dropout(0.5) (Dense2)
PDense2 = PReLU() (Dense2)


CTR = Dense(2, activation='softmax', use_bias = False, name= 'ctr') (PDense2)

model = Model(inputs=[userInput, prodInput, candProdInput], outputs= [CTR])
#opt = tf.keras.optimizers.Adam(learning_rate=0.0001)
model.compile(optimizer= 'adam',
              loss='binary_crossentropy',
              metrics=['binary_accuracy'])

checkpointer = ModelCheckpoint(filepath = modelName+'.weights.best.hdf5', verbose=1, save_best_only = True)
#model.fit(X_train,y_train,epochs = 20, batch_size = 8)

model.fit([userId_train, prodId_train, candProd_train], [candRate], epochs = 1 ,callbacks=[checkpointer], shuffle= True, validation_split=0.2, batch_size = 200)


model.save(modelName+'.h5')
```
